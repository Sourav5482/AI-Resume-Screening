```python

# def extract_resume_text(pdf_file):
#     text = ""
#     with pdfplumber.open(pdf_file) as pdf:
#         for page in pdf.pages:
#             if page.extract_text():
#                 text += page.extract_text() + " "
#     return text
# resume_text = extract_resume_text(file_name)
# print(resume_text)
import pdfplumber
import io

def extract_resume_text(upload_widget):
    text = ""

    uploaded = upload_widget.value

    # ipywidgets v8 â†’ tuple
    if isinstance(uploaded, tuple):
        uploaded_file = uploaded[0]
        file_bytes = uploaded_file["content"]

    # ipywidgets v7 â†’ dict
    else:
        uploaded_file = next(iter(uploaded.values()))
        file_bytes = uploaded_file["content"]

    with pdfplumber.open(io.BytesIO(file_bytes)) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + " "

    return text

```

```python
resume_text = extract_resume_text(file_name)
print(resume_text)
```

```python
# llm = pipeline(
#     "text2text-generation",
#     model="google/flan-t5-base",
#     max_length=120,
#     do_sample=False
# )
from transformers import pipeline

llm = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    framework="pt",     # ðŸ‘ˆ IMPORTANT
    max_length=120,
    do_sample=False
)
```

```python

job_description = """
We are hiring for a Data Analyst role.
Required skills:
- Python
- SQL
- Data Analysis
- Communication
- Problem Solving
"""

prompt = f"""
You are an AI resume screening system.

Compare the resume with the job description and classify.

Job Description:
{job_description}

Resume:
{resume_text}

Return EXACTLY this format (no extra text):

Matched Skills: <skills>
Match Percentage: <number>%
Final Decision: <Suitable or Not Suitable>
"""
```

```python

decision = llm(prompt)[0]["generated_text"].strip()
skill_bank = [
    "python", "sql", "data analysis", "javascript", "react",
    "git", "html", "css", "mongodb", "communication"
]

matched_skills = [
    skill.title() for skill in skill_bank if skill in resume_text.lower()
]

match_percentage = min(len(matched_skills) * 10, 100)
print("ðŸ¤– AI Resume Screening Result\n")
print(f"Matched Skills: {', '.join(matched_skills) if matched_skills else 'None'}")
print(f"Match Percentage: {match_percentage}%")
print(f"Final Decision: {decision}")
```
